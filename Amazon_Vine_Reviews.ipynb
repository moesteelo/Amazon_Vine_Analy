```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change the path if needed.)
vine_summary_to_load = "vine.csv"

# Read the Amazon Vine Reviews into a Pandas DataFrame
vine_summary_df = pd.read_csv(vine_summary_to_load, low_memory=False)

# Check Data Frame
vine_summary_df.head(5)
```

```python
total_votes_summary = vine_summary_df[vine_summary_df["total_votes"] >= 20]
total_votes_summary.head()
```

```python
top_vine_reviews = total_votes_summary[total_votes_summary["helpful_votes"]/total_votes_summary["total_votes"] >= .50]
top_vine_reviews.head()
```

```python
paid_vine_5star = top_vine_reviews[top_vine_reviews["vine"] == 'Y']
paid_vine_5star.head()
```

```python
nonpaid_vine_5star = top_vine_reviews[top_vine_reviews["vine"] == 'N']
nonpaid_vine_5star.head()
```

```python
total_paid_vine_reviews = paid_vine_5star.count()[0]
# total_paid_vine_reviews
```

```python
vine_5star_rating = paid_vine_5star[paid_vine_5star["star_rating"] == "5"].count()[0]
#vine_5star_rating.head()
```

```python
paid_5star_percent = (vine_5star_rating/total_paid_vine_reviews) * 100
# paid_5star_percent
```

```python
total_nonpaid_vine_reviews = nonpaid_vine_5star.count()[0]
# total_nonpaid_vine_reviews
```

```python
non_vine_5star_rating = nonpaid_vine_5star[nonpaid_vine_5star["star_rating"] == "5"].count()[0]
non_vine_5star_rating
```

```python
non_paid_5star_percent = (non_vine_5star_rating/total_nonpaid_vine_reviews) * 100
non_paid_5star_percent
```

```python
amazon_vine_reviews = pd.DataFrame({'Total_vine_reviews':[total_paid_vine_reviews],
                            'Total_non_vine_reviews':[total_nonpaid_vine_reviews],
                            'Total_5_star_vine_reviews':[vine_5star_rating],
                            'Total_5_star_non_vine_reviews':[non_vine_5star_rating],
                            'Vine_reviews_5_star_percentage':[paid_5star_percent],
                            'Non_vine_reviews_5_star_percentage':[non_paid_5star_percent]})

amazon_vine_reviews["Vine_reviews_5_star_percentage"] = amazon_vine_reviews["Vine_reviews_5_star_percentage"].map("{:.2f}%".format)
amazon_vine_reviews["Non_vine_reviews_5_star_percentage"] = amazon_vine_reviews["Non_vine_reviews_5_star_percentage"].map("{:.2f}%".format)
```
